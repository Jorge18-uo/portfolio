```
def dibujar_escalera(n):
    if n == 0:
        print("__")
    else:
        for i in range(abs(n)):
            if n > 0:
                print(" "*(n-i-1) + "_"*(i+1) + "|")
            else:
                print(" "*i + "|" + "_"*(abs(n)-i-1))
```

```
def dibujar_escalera():
    continuar = True
    while continuar:
        n = int(input('Ingrese la cantidad de escalones a agregar (o ingrese 0 para detenerse): '))
        if n == 0:
            continuar = False
        elif n > 0:
            for i in range(n):
                print(" "*(n-i-1) + "_"*(i+1) + "|")
        else:
            for i in range(abs(n)):
                print(" "*i + "|" + "_"*(abs(n)-i-1))
                
```
